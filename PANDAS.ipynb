# Removing Duplicates

```python
data.duplicated()
```

```python
data.drop_duplicates() # drops rows with same values in all columns
```

```python
data.drop_duplicates('col 3') #drop rows with same values in specified column  
```

# Grouping

```python
cars_gear = cars.groupby(cars.gear) #gear 3,4,5 is present in dataset,
cars_gear.mean() #mean values of particular group 
```

```python
cars_am = cars.groupby(cars.am)
cars_am.mean()
```

```python
cars_groups = cars.groupby([cars.gear,cars.am])
cars_groups.mean()
```
