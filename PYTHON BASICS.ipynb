## 6. Loops:
#### For loop:

```python
flowers = ['lotus','rose','orchild',
          'sunflower','jasmine','lilly',
          'tulip','hibiscus']
```

```python
print("the length of", flowers[0], "is", len(flowers[0]))
```

```python
print("the length of", flowers[1], "is", len(flowers[1]))
print("the lenght of", flowers[2], "is", len(flowers[2]))
print("the lenght of", flowers[3], "is", len(flowers[3]))
print("the length of", flowers[4], "is", len(flowers[4]))
print("the lenght of", flowers[5], "is", len(flowers[5]))
      
```

```python
for t in flowers:
    print("the lenght of", t, "is", len(t)) #here 'f' denotes flowers
    
```

```python
for i,f in enumerate(flowers):
    print(i,"the lenght of", f, "is", len(f)) #i gives the serial number indexing we use enumerate to know the index number.
```

```python
fruits=['apple', 'mango','guava', 'jack fruit',]
```

```python
print("the numbers of characters present in the",fruits[0], "is", len(fruits[0]))
```

```python
print("the", fruits[1], "has", len(fruits[1]),"as number of characters") 
print("the", fruits[2], "has", len(fruits[2]),"as number of characters")
```

```python
for f in fruits:
    print("the", f, "has", len(f), "as number of characters")
```

## Treating missing values in python

```python
cars.loc[5,'hp']=np.nan
cars.loc[3,'hp']=np.nan
cars.loc[6,'hp']=np.nan
cars.loc[5,'mpg']=np.nan
cars.loc[2,'mpg']=np.nan
cars.head(10)
```

```python
cars.isnull().sum()
```

## Filling the missing values with fillna()

```python
cars.fillna(100)
```

```python
# Treating missing values using forward filling, i.e filling the above value
cars.fillna(method='ffill')
```

```python
#Treating missing values using backward filling , i.e filling the value by below value.
cars.fillna(method='bfill')
```

```python
#Treating missing values with the mean. it gives better reading.
cars.fillna(cars.mean())
```

## Removing rows with missing values

```python
cars.dropna()
```

```python
cars.head() #but still nan appears 
```

```python
cars.dropna(inplace=True) #drops the missing row and replaces the original value.
cars.head(10)
```

```python
cars.dropna(how='all') # only drop row when all columns of that row is empty
```
